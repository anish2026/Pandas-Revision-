```python
# now back to our original dataset

df=pd.read_csv('survey_results_public.csv',index_col='Respondent')
schema=pd.read_csv('survey_results_schema.csv',index_col='Column')
pd.set_option('display.max_rows',85)

#now applying some filter , lets check salary >30000, for that we can look schema salary column exists or not
# so we came through that salary is with name 


schema
```

```python
# ConvertedComp so we found that this is the salary , now we will make filter for that

salary_filter = (df['ConvertedComp'] > 300000)
df.loc[salary_filter] 

#suppose we want to concentrate on salary only 

df.loc[salary_filter,'ConvertedComp']
```

```python
#now we want to filter my Indians brothers and sisters with salary greater than 500000
#defining a new filter

new_filter = (df['ConvertedComp'] > 500000) & (df['Country']=='India')
df.loc[new_filter,['ConvertedComp','Country']]
```

```python
#use of isin() method

#now we wqant to filter all those whose salary is greater than 400000 and are from India,United States,Canada,Ukraine

countries = ['India','United States','Canada','Ukraine']

filterrr = (df['ConvertedComp'] > 400000) & (df['Country'].isin(countries))
df.loc[filterrr,['ConvertedComp','Country']]
```

```python
#use of str.contains()

#now we are interested in knowing the Programming languages Python is used by how much people
#here is a problem , in out dataset the Languages are seperated with comma or columns and represented as string 

df['LanguageWorkedWith']
```

```python
#we need to filter only python from this , so we will make a filter with name lang

lang = df['LanguageWorkedWith'].str.contains('Python')
df.loc[lang]
```

#### In the above we got , ValueError: Cannot mask with non-boolean array containing NA / NaN values

This is because there are some rows in which languages worked with are NaN i.e Bool type , so str.contains is throwing value error.

We can solve this problem by filtering NaN values , passing an argument na=False in contains method

```python
lang = df['LanguageWorkedWith'].str.contains('Python',na=False)    #rewriting my code 
df.loc[lang,['LanguageWorkedWith']]
```
